```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
data = pd.read_csv('./data/processed_data.csv',sep='\t')
```

```python
data.head()
```

```python
US = data[data['Country'] == 'United States']
US.head()
unique_attributes = US['ATTRIBUTE'].unique()
grouped = US.groupby('ATTRIBUTE')
grouped.head()
```

```python
#extract for year 2004 only
US_4 = US[US['TIME_PERIOD'] == 2004]
#remove Country 
US_4 = US_4.drop('Country', axis=1)
#replace TIME_PERIOD with year, OBS_VALUE with obser_value, and ATTRIBUTE with attribute
US_4 = US_4.rename(columns={'TIME_PERIOD':'year','OBS_VALUE':'observe_value','ATTRIBUTE':'attribute'})
US_4.head()
#flat into single row
US_4 = US_4.pivot(index='year', columns='attribute', values='observe_value')
US_4.tail()
#Do the same for all years
US_all = US.drop('Country', axis=1)
US_all = US_all.rename(columns={'TIME_PERIOD':'year','OBS_VALUE':'observe_value','ATTRIBUTE':'attribute'})
US_all = US_all.pivot(index='year', columns='attribute', values='observe_value')
US_all.head()
```

```python
#print all SAD_deaths columns
sad_columns = [col for col in US_all.columns if 'SAD_deaths' in col]
print(sad_columns)
#put SAD_deaths_Age:Total_Sex:Female_Education:Total, SAD_deaths_Age:Total_Sex:Male_Education:Total and SAD_deaths_Age:Total_Sex:Total_Education:Total in the end of the columns in US_all
US_all = US_all[[col for col in US_all.columns if col not in sad_columns] + sad_columns]
US_all.head()
#save US_all to a csv file
US_all.to_csv('./data/US_all.csv')
```

```python
#Repeat the same for all countries
countries = data['Country'].unique()
for country in countries:
    country_data = data[data['Country'] == country]
    country_data = country_data.drop('Country', axis=1)
    country_data = country_data.rename(columns={'TIME_PERIOD':'year','OBS_VALUE':'observe_value','ATTRIBUTE':'attribute'})
    country_data = country_data.pivot(index='year', columns='attribute', values='observe_value')
    sad_columns = [col for col in country_data.columns if 'SAD_deaths' in col]
    country_data = country_data[[col for col in country_data.columns if col not in sad_columns] + sad_columns]
    country_data.to_csv('./data/'+country+'_all.csv')
```
